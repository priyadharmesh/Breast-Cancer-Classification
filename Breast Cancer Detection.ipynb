```python
#500 Training

X_train=feature[:500]
Y_train=label[:500]

#35 Validation

X_val=feature[500:535]
Y_val=label[500:535]


#34 Testing

X_test=feature[535:]
Y_test=label[535:]



```

```python
X_train.shape
```

```python
#neural network
```

```python
model = Sequential()

model.add(Dense(15, activation = 'relu' , input_dim = 30))  #each tumor is having 30 values as input and one value as output

# activation function :it tells whether the neuron should get activated or not

model.add(Dense(15, activation = 'relu'))


model.add(Dense(15, activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))
#activation fun:sigmoid becuase we want the values between 1 and 0
#dense:neural with one dimension

model.compile( loss='binary_crossentropy' , optimizer='adam' , metrics=['accuracy'])
```

```python
model.fit(X_train , Y_train , batch_size=1, epochs = 5 , validation_data=(X_val,Y_val))
```

```python
model.evaluate(X_test , Y_test)
```
